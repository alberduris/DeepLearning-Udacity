```python
"""Softmax."""

#scores = [3.0, 1.0, 0.2] --> scores = [30.0, 10.0, 2]

import numpy as np

def softmax(x):
    """Compute softmax values for each sets of scores in x."""
    return np.exp(x) / np.sum(np.exp(x),axis=0)

# Plot softmax curves
import matplotlib.pyplot as plt
x = np.arange(-2.0, 6.0, 0.1)
scores = np.vstack([x, np.ones_like(x), 0.2 * np.ones_like(x)])

plt.plot(x, softmax(scores*10).T, linewidth=2)
plt.show()
```

```python
#Lo que ocurre es que las probabilidades se convierten en probabilidades muy cercanas a 0 ó 1
```

```python
"""Softmax."""

#scores = [3.0, 1.0, 0.2] --> scores = [ 0.3   0.1   0.02]

import numpy as np

def softmax(x):
    """Compute softmax values for each sets of scores in x."""
    return np.exp(x) / np.sum(np.exp(x),axis=0)

# Plot softmax curves
import matplotlib.pyplot as plt
x = np.arange(-2.0, 6.0, 0.1)
scores = np.vstack([x, np.ones_like(x), 0.2 * np.ones_like(x)])

plt.plot(x, softmax(scores/10).T, linewidth=2)
plt.show()
```

```python
#Lo que ocurre es que las probabilidades se convierten en una probabilidad uniforme
#since all the scores decrease in magnitude, the resulting softmax probabilities will be closer to each other.
```
